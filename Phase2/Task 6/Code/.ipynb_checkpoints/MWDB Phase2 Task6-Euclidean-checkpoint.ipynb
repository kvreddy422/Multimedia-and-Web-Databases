```python
# Creating the Symmetry matrix
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine

dist_out = pairwise_distances(df.iloc[:,1:], metric="euclidean")
dist_out_norm = (dist_out - dist_out.min()) / (dist_out.max() - dist_out.min())
sim_df = pd.DataFrame(dist_out_norm)
```

```python
# SVD done on the input Matrix
from scipy import linalg
U, sigma, V = np.linalg.svd(sim_df)
```
