```python
# Getting the Similarity Square Matrix
from scipy.spatial.distance import squareform, pdist
df
[rows,columns]=df.shape
rows
columns
# df.iloc[:,1:]
dist = pdist(df.iloc[:,1:], metric="cosine")
s_dist = squareform(dist)
s_dist.shape
# # Fill diagonals with nulls
np.fill_diagonal(s_dist, 0)
# s_dist_norm = (s_dist - s_dist.mean()) / (s_dist.max() - s_dist.min())
sim = np.subtract(1, s_dist)
sim_df = pd.DataFrame(sim)
# sim_df=sim_df.fillna(1)
sim_df
```

```python
# SVD done on the input Matrix
from scipy import linalg
U, sigma, V = np.linalg.svd(sim_df)
```
